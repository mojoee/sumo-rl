```python
import pandas as pd
import matplotlib.pyplot as plt

# Paths to your CSV files
file_paths = [
    'finalOutput/B/125/summary_run1seed1111.csv',
    'finalOutput/B/125/summary_run1seed2222.csv',
    'finalOutput/B/125/summary_run1seed3333.csv',
    'finalOutput/B/125/summary_run1seed4444.csv',
    
    'finalOutputNormal/B/125/summary_run1_seed1110.csv',
    'finalOutputNormal/B/125/summary_run1_seed2202.csv',
    'finalOutputNormal/B/125/summary_run1_seed3033.csv',
    'finalOutputNormal/B/125/summary_run1_seed4404.csv',

]


# Metrics to analyze
metrics = ['step_meanTravelTime', 'step_meanWaitingTime']  # Added step_meanWaitingTime

# Initialize a dictionary to store statistics
stats = {metric: [] for metric in metrics}
max_arrived = []

# Loop through files and calculate statistics
for file_path in file_paths:
    df = pd.read_csv(file_path, sep=';')
    
    # Calculate statistics for specified metrics
    for metric in metrics:
        stats[metric].append(df[metric].describe())
    
    # Calculate and store the maximum value of step_arrived
    max_arrived.append(df['step_arrived'].max())

# Print statistics for each metric
for metric, values in stats.items():
    print(f"Statistics for {metric}:")
    for i, value in enumerate(values):
        column = pd.read_csv(file_paths[i], sep=';')[metric]
        print(f"Run {i + 1}:\n{value}")
        print("Additional Statistics:")
        print(f"Median: {column.median()}")
        print(f"Mode: {column.mode()[0]}")
        print(f"Standard Deviation: {column.std()}\n")

# Print maximum values of step_arrived
for i, max_val in enumerate(max_arrived):
    print(f"Maximum 'step_arrived' for Run {i+1}: {max_val}")

# Visualization: Box plot for each metric
for metric in metrics:
    data = [pd.read_csv(file, sep=';')[metric] for file in file_paths]
    plt.figure()
    plt.boxplot(data)
    plt.title(f'Box Plot of {metric}')
    plt.xticks(range(1, len(file_paths) + 1), [f'Run {i+1}' for i in range(len(file_paths))])
    plt.ylabel(metric)
    plt.show()

import pandas as pd
import matplotlib.pyplot as plt

file_paths = [
    'finalOutputNormal/A/75/tripinfo_run1_seed1111.csv',
    'finalOutputNormal/A/75/tripinfo_run1_seed2222.csv',
    'finalOutputNormal/A/75/tripinfo_run1_seed3333.csv',
    'finalOutputNormal/A/75/tripinfo_run1_seed4444.csv',

    'finalOutput/A/75/tripinfo_run1seed1111.csv',
    'finalOutput/A/75/tripinfo_run1seed2222.csv',
    'finalOutput/A/75/tripinfo_run1seed3333.csv',
    'finalOutput/A/75/tripinfo_run1seed4444.csv',
]

# Metric to analyze
metric = 'tripinfo_duration'

average_durations = {}

for file_path in file_paths:
    df = pd.read_csv(file_path, sep=';')

    # Group by vehicle type and calculate the average trip duration
    grouped = df.groupby('tripinfo_vType')
    average_durations[file_path] = {}

    for vType, group in grouped:
        average_duration = group[metric].mean()  # Average duration for this vehicle type
        average_durations[file_path][vType] = average_duration

# Print average durations for each vehicle type in each dataset
for file_path, vTypes in average_durations.items():
    print(f"Average trip duration for each vehicle type in {file_path}:")
    for vType, avg_duration in vTypes.items():
        print(f"{vType}: {avg_duration}")

# Visualization: Bar plot for average trip duration for each vehicle type in each dataset
for file_path in file_paths:
    vehicle_types = list(average_durations[file_path].keys())
    avg_durations = [average_durations[file_path][vType] for vType in vehicle_types]

    plt.figure()
    plt.bar(vehicle_types, avg_durations)
    plt.title(f'Average Trip Duration for Each Vehicle Type in {file_path}')
    plt.xlabel('Vehicle Type')
    plt.ylabel('Average Trip Duration')
    plt.show()
```
