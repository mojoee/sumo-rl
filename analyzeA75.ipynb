```python
import pandas as pd
import matplotlib.pyplot as plt

# Read your CSV file
file_path = ('finalOutputNormal/A/75/tripinfo_run1.csv')
df = pd.read_csv(file_path, sep=';')

# Select the column
column_name = 'tripinfo_duration'  # Replace with the name of your column
column = df[column_name]

# Descriptive statistics for numerical data
if column.dtype == 'float64' or column.dtype == 'int64':
    column_sum = column.sum()

    print("Descriptive Statistics:")
    print(column.describe())
    print(f"Sum of {column_name}: {column_sum}")

    # Additional statistics
    print("\nAdditional Statistics:")
    print(f"Median: {column.median()}")
    print(f"Mode: {column.mode()[0]}")
    print(f"Standard Deviation: {column.std()}")

    # Summation Plot
    plt.figure()
    column.cumsum().plot(kind='line')
    plt.title('Cumulative Sum over Interval')
    plt.xlabel('Index')
    plt.ylabel('Cumulative Sum')
    plt.show()

    # Box Plot
    plt.figure()
    column.plot(kind='box')
    plt.title('Box Plot')
    plt.ylabel(column_name)
    plt.show()

    # Histogram
    plt.figure()
    column.plot(kind='hist', bins=30)
    plt.title('Histogram')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.show()

# Frequency counts for categorical data
if column.dtype == 'object':
    print("Frequency Counts:")
    print(column.value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Paths to your CSV files
file_paths = [
    'finalOutputNormal/A/75/tripinfo_run1.csv',
    'finalOutputNormal/A/75/tripinfo_run2.csv',
    'finalOutputNormal/A/75/tripinfo_run3.csv',
    'finalOutputNormal/A/75/tripinfo_run4.csv',
    'finalOutputNormal/A/75/tripinfo_run5.csv'
]

# Column name to analyze
column_name = 'tripinfo_duration'  # Replace with the name of your column

# Initialize a dictionary to store DataFrames
dataframes = {}
stats = {}

for i, file_path in enumerate(file_paths):
    # Read each file
    df = pd.read_csv(file_path, sep=';')
    dataframes[f'File {i+1}'] = df

    # Calculate descriptive statistics
    column = df[column_name]
    stats[f'File {i+1}'] = column.describe()
    stats[f'File {i+1}']['Sum'] = column.sum()
    stats[f'File {i+1}']['Median'] = column.median()
    stats[f'File {i+1}']['Mode'] = column.mode()[0]
    stats[f'File {i+1}']['Standard Deviation'] = column.std()

# Print statistics
for file, stat in stats.items():
    print(f"Statistics for {file}:")
    print(stat)
    print("\n")

fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(10, 20))
fig.suptitle('Histograms for Each File')

for i, (file, df) in enumerate(dataframes.items()):
    df[column_name].plot(kind='hist', bins=30, ax=axes[i], title=f"Histogram for {file}")
    axes[i].set_xlabel(column_name)
    axes[i].set_ylabel('Frequency')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
